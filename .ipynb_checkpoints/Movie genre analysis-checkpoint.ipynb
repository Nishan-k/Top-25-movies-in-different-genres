### Importing the libraries

```python
from urllib.request import Request, urlopen
import pandas as pd
import seaborn as sns
import urllib.request
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import re
```

### Scraping the data and Converting it to a dataframe

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
base_url = "https://www.the-numbers.com/market/genres"
req = Request(url=base_url, headers=headers) 
html = urlopen(req).read() 
bs_obj = soup(html)
```

```python
headings = bs_obj.find_all('th')
```

```python
table_body = bs_obj.find_all("td")
```

```python
raw_data = []
for i in range(0, len(table_body)):
    raw_data.append( table_body[i].get_text())
```

```python
data = []
i = 0
j = 6
for _ in range(0, 15):
    data.append(raw_data[i:j])
    i+=6
    j+=6
```

```python
headings = [heading.get_text() for heading in headings]
df_movies_genre = pd.DataFrame(data, columns=headings, index=None )
```

```python
df_movies_genre
```

```python
df_movies_genre.info()
```

```python
df_movies_genre = df_movies_genre.replace(',','', regex=True)
df_movies_genre['Movies'] = df_movies_genre['Movies'].astype(int)
```

### Visualizing the number of movies produced in each genre between 1995 to 2022

```python
plt.figure(figsize = (18,10))
ax = sns.barplot(x = 'Genre', y = 'Movies',data = df_movies_genre)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.title("Number Of Movies In Different Genres (1995 - 2022)",fontdict={'fontsize':14})
plt.xlabel(xlabel='Genre',fontdict={'fontsize':14})
plt.ylabel(ylabel='Movies',fontdict={'fontsize':14})
ax.bar_label(ax.containers[0])
plt.show()
```

### Building a crawler:

Now, we will crawl through each of the genre and analyse the data

```python
td_tags = bs_obj.find_all('td')
```

```python
links  = [] 
for td in td_tags:
    if (td.find('b')) != None:
        links.append('https://www.the-numbers.com/' + (td.find('a')['href']) )
```

```python
links
```

```python
raw_movie_data = []
```

```python
def crawler():
    for i in range(0, len(links)):
        req = Request(url=links[i], headers=headers) 
        html = urlopen(req).read()
        bsObj = soup(html)
        first_table = bsObj.find("table")
        table_rows = first_table.find_all("tr")
        raw_movie_data.append(list(table_rows))
```

```python
crawler()
```

```python
len(raw_movie_data)
```

```python
clean_data = [adventure_data,action_data,drama_data,comedy_data,thriller_data,horror_data,romantic_comedy_data,musical_data,documentary_data,black_comedy_data,western_data,concert_data,multiple_genre_data ]
```

### Cleaning the dataset except for reality and educational genre:

```python
for i in range(0, len(raw_data)):
    for j in range(1, len(raw_data[0])-1):
        raw_text = raw_data[i][j].get_text().split('\n')
        clean_data[i].append(list(filter(lambda lines: lines, raw_text)))
```

### Data Cleaning: Reality Genre Data


```python
reality_data =[]
for i in range(1, len(reality_raw)-1):
    raw_text = reality_raw[i].get_text().split('\n')
    reality_data.append(list(filter(lambda lines: lines, raw_text)))
reality_data[-1].insert(3,'Not Available')
```

### Data Cleaning: Educational Genre Data


```python
educational_data =[]
for i in range(1, len(educational_raw)-1):
    raw_text = educational_raw[i].get_text().split('\n')
    educational_data.append(list(filter(lambda lines: lines, raw_text)))

educational_data[0].insert(4,'Not Rated')
educational_data[0][1] = (re.sub(r'[â\x80\x99]+','"', educational_data[0][1]))
```

### Creating Dataframes:

```python
headings = ['Rank', 'Movie','Release Date','Theatrical Distributor','MPAA(Motion Picure)Rating','Total Gross','Inflation-Adjusted Gross']
```

```python
adventure_data = list()
action_data = list()
drama_data = list()
comedy_data = list()
thriller_data = list()
horror_data = list()
romantic_comedy_data = list()
musical_data = list()
documentary_data = list()
black_comedy_data = list()
western_data = list()
concert_data = list()
multiple_genre_data = list()
```

```python
df_adventure = pd.DataFrame(clean_data[0],columns=headings, index = None)
df_action = pd.DataFrame(clean_data[1],columns=headings, index = None)
df_drama = pd.DataFrame(clean_data[2],columns=headings, index = None)
df_comedy = pd.DataFrame(clean_data[3],columns=headings, index = None)
df_thriller = pd.DataFrame(clean_data[4],columns=headings, index = None)
df_horror = pd.DataFrame(clean_data[5],columns=headings, index = None)
df_romantic_comedy= pd.DataFrame(clean_data[6],columns=headings, index = None)
df_musical = pd.DataFrame(clean_data[7],columns=headings, index = None)
df_documentary = pd.DataFrame(clean_data[8],columns=headings, index = None)
df_black_comedy = pd.DataFrame(clean_data[9],columns=headings, index = None)
df_western = pd.DataFrame(clean_data[10],columns=headings, index = None)
df_concert = pd.DataFrame(clean_data[11],columns=headings, index = None)
df_multiple_genre = pd.DataFrame(clean_data[12],columns=headings, index = None)
df_reality = pd.DataFrame(reality_data, columns = headings, index = None)
df_educational = pd.DataFrame(educational_data, columns=headings, index= None)
```

```python
df_adventure['Genre'] = ['Adventure' for i in range(0, len(df_adventure))]
df_action['Genre'] = ['Action' for i in range(0, len(df_action))]
df_drama['Genre'] = ['Drama' for i in range(0, len(df_drama))]
df_comedy['Genre'] = ['Comedy' for i in range(0, len(df_comedy))]
df_thriller['Genre'] = ['Thriller' for i in range(0, len(df_thriller))]
df_horror['Genre'] = ['Horror' for i in range(0, len(df_horror))]
df_romantic_comedy['Genre'] = ['Romantic Comedy' for i in range(0, len(df_romantic_comedy))]
df_musical['Genre'] = ['Musical' for i in range(0, len(df_musical))]
df_documentary['Genre'] = ['Documentary' for i in range(0, len(df_documentary))]
df_black_comedy['Genre'] = ['Black Comedy' for i in range(0, len(df_black_comedy))]
df_western['Genre'] = ['Western' for i in range(0, len(df_western))]
df_concert['Genre'] = ['Concert' for i in range(0, len(df_concert))]
df_multiple_genre['Genre'] = ['Multiple Genre' for i in range(0, len(df_multiple_genre))]
df_reality['Genre'] = ['Reality' for i in range(0, len(df_reality))]
df_educational['Genre'] = ['Educational']
```

### Final Dataframe of all the genres into a single one:

```python
movie_genre = pd.concat([df_adventure,df_action,df_drama,df_comedy,df_thriller,df_horror,df_romantic_comedy,df_musical,df_documentary,df_black_comedy,df_western,df_concert,df_multiple_genre,df_reality,df_educational], ignore_index=True)
```
